```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Salaries.csv')
df.head()
```

```python
# Identify the number of rows and columns
print('Columns')
print(df.columns)
print('\n Numbers of Rows and Columns')
print('Number of Rows: ',df.shape[0])
print('Number of Columns: ',df.shape[1])
```

```python
# Determine the data types of each column
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Drop Notes, Status Columns Because are Contains more than 99% of missing values and some not nescessary
df.drop(columns=['Id','Notes', 'Agency','Status'], inplace=True)
df.head()
```

```python
# calucate missing values of datasets
total_missing = df.isnull().sum()
total_rows = df.shape[0]

(total_missing/total_rows) * 100
```

```python
# fill mean of Benefits Column Because it's contains about 24% of total dataset rows and delete this rows will change result
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

```python
# Drop all null values Because have less than 1% of dataset missing value and this don't change dataset result
df.dropna(inplace=True)
```

```python
# check the missing values
total_missing
```

```python
df.head()
```

# Calculate basic statistics


```python
# Calculate Mean
columnsOfSalary = ['BasePay','OvertimePay','OtherPay','Benefits', 'TotalPay', 'TotalPayBenefits']
df[columnsOfSalary].mean()
```

```python
# Calculate Median
df[columnsOfSalary].median()
```

```python
# Calculate Mode
df[columnsOfSalary].mode()
```

```python
# Calculate Min
df[columnsOfSalary].min()
```

```python
# Calculate Max
df[columnsOfSalary].max()
```

```python
# Calculate Standard Deviation
df[columnsOfSalary].std()
```

```python
# Calculate Range Of Data
def range1(series):
  return series.max() - series.min().abs()

range1(df[columnsOfSalary])
```
